```python
import numpy as np
import matplotlib.pyplot as plt

from kernel.fun import KernelFun, KernelFun2
from kernel.rect import KernelRect

%load_ext autoreload
%autoreload 2
```

```python
# tbins = np.arange(20, 100, 5)

fig, ax = plt.subplots()
t = np.arange(0, 200., .1)
stim = np.zeros((len(t), 1))
stim[500] = 1
ax.plot(t, stim * 0.5, 'k-')

for t0 in np.arange(-25, 75, 25):
    tbins = np.arange(t0, t0 + 105, 5)
    coefs = 1e1 * np.exp(-(tbins[:-1] - tbins[0])/10.)
    ker = KernelRect(tbins, coefs)
    y = ker.convolve_continuous(t, stim)
    ax.plot(t, y, '-')
```
